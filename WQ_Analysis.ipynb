```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats 
import seaborn as sns
```

```python
data = pd.read_csv("C:/Users/Administrator/Desktop/AI & ML/Analysis/analysis.csv")
```

```python
data.info()
```

```python
data.describe().round(3).to_csv("descriptive.csv")
```

```python
scatter_plot = sns.pairplot(data.loc[:, 'pH':'Sulphate'], height = 5)
scatter_plot
```

```python
#Histogram plot using matplotlib for pH

plt.hist(data['pH'], bins = 10, color = 'blue')
plt.xlabel("pH")
plt.title("Histogram of pH value")
plt.show
plt.savefig('pH hist.jpg', bb_box = 'tight')
```

```python
#Boxplot of pH using matplotlib 
plt.boxplot(data['pH'])
plt.show()
```

```python
hcf = data[data['Type'] == 'HCF']
school = data[data['Type'] == 'School']
ea = data[data['Type'] == 'EA']

plt.boxplot([hcf['pH'], school['pH'], ea['pH']])
plt.ylabel('pH')
plt.xlabel('HCF, School, EA')
plt.title("Boxplot of pH based on different water users")
plt.savefig('Boxplot_pH.jpg')
```
